## Introduction and Motivation 
~ [Q1] - [Q9]

[Q1] The main motivation for the paper focuses on broadening our understanding of the direct and indirect impacts of assigning firms to public or private citizen appeals when they violate pollution standards in China. 

[Q2] The authors seek to explain the following questions. 1) Can public appeals to regulators via social media significantly reduce firms' environmental violations and emissions? 2) What is the impact of private appeals to regulators on corporate environmental violations and emissions? 3) Will these appeals have a spillover effect on the firms' emissions, that is, will the public appeals cause other firms not directly complained about to also reduce their emissions?

[Q3] First, through field experiments across the country, the authors assess the impact of citizens on the behavior of polluting companies and environmental governance through public and private appeals channels. They examine not only the direct effects of citizen appeals, but also the possible indirect effects of citizen appeals, namely whether they change the pollution behavior of non-targeted firms and entire regions. Second, they assess the impacts of social media on environmental governance.  Although previous literature has covered its effects on citizens’ political behavior and attitude, few articles have addressed the role of social media in environmental governance. This article fills a gap in empirical research on the role of citizen participation in environmental governance.

[Q4] This article firstly connects citizen participation, regulatory effort, and adjustments by firms, showing how participation can meaningfully decrease pollution. It also relates to many fields, including political economy, citizen behavior, and government structure. Through large-scale field experiments and detailed data analysis, this paper provides new insights and empirical support, expanding the research boundaries in the field of environmental political economy and public governance.

[Q5] In the context of China's environmental governance, the institutional environment can be described as a combination of "top-down" and "bottom-up".  The Chinese government set specific environmental performance targets , launched nationwide pollution monitoring networks and provided real-time pollution information, which are “top-down” measures. Additionally, the government also encourages citizens to report violations of pollution standards via the hotline “12369”. They can call the hotline or leave a message on the website. Their appeals will be directed to the corresponding local EPA. 


### Firm characteristics

[Q6] Not all polluting firms are monitored in China; in the context of this paper, the firms that are monitored are the 24,620 major polluting plants, which are monitored by the Continuous Emissions Monitoring System (CEMS) arm of the Ministry of Ecology and the Environment (MEE). These firms are the most important to monitor; the paper describes that they are responsible for 75 percent of China’s total industrial emissions. A firm is considered a key polluter if their total emissions of the air pollutants $SO_2$, PM, $NO_X$ or the water pollutants $COD$ or $N_H3-N$ have in the past two years, exceeded the threshold that is determined by the MEE. 

### Enforcement and Regulation

[Q7] The enforcement of environmental standards depends largely on the CEMS, which is a network of automatic monitoring equipment consisting of apparatuses and flow meters that are installed at the discharge points. Using the CEMS data, compliance is enforced by the local EPA and is carried out as follows. First, if CEMS data indicate violation by a firm, the local EPA must verify the violation before they may inflict penalties or order shutdown. The agency then issues a warning and can only impose penalties if standards are not met one month following this initial warning, which would be found in the subsequent inspection. The benefit of this approach is that firms are given the opportunity to remedy their pollution violation during the one-month window, which guides them towards more efficient production practices. Furthermore, the threat of penalties inflicted by the local agency is a motivating incentive that ensures the firms will adhere to the pollution standards. The drawback of this approach, as explained in the paper is the conflicting incentives that may arise between the local and central government. While the central government imposes standards that apply across prefectures, the local agencies face a tradeoff between productivity and pollution. In these prefectures, there is higher productivity and economic growth associated with more pollution, and at this small scale, little harm done in more pollution given that it does not create public discourse, so they have “exercised discretion in levying penalties against violators”.

### Regulator objectives; conflicting incentive and moral hazard

[Q8] The two main objectives that the regulators are trying to balance between are meeting the environmental objectives set by the central government and the economic outcomes of productivity and prosperity. There is a tradeoff in these objectives, because higher productivity corresponds to increased pollution by the firms. The paper explains how these objectives conflict. Focusing on economic growth, political leaders have influence on the ways in which environmental regulators’ careers advance and have emphasized the importance of economic growth. There are instances in which local regulators were given orders that their regulations would not affect production and employment. Shifting towards the environmental side, the regulators face the equally challenging threat of social unrests and protest in response to high levels of pollution.

The following figure illustrates the trend for percentage of violations for air pollutants $SO_2$, PM, $NO_X$ and the water pollutants $COD$ or $N_H3-N$ in the CEMS data spanning 2018 to 2021.  

[Q9] This figure illustrates the moral hazard issue that [Q8] alluded to. 
One potential reason for the discontinuity and the substantial increase in violation rate for water pollution and the slight increase in that of gas pollution is election cycles. As described in Q8, the **local regulators face the conflicting incentives of environmental regulation compliance and the economic growth associated with lax regulation**. Towards the end of 2018, it is possible that upcoming elections influenced the politicians, whom the local regulators’ career trajectories depend on, so they levied fewer penalties, as this would yield higher production and employment. In an economic sense, this would cast the politicians in a favorable light so they can get re-elected. Not only does this constitute a moral hazard problem, but also potentially explains the discontinuity; prior to the election, in order to boost the economy, more violations of the air and water pollution standards occur. On January 1, following the election, the incumbents have re-secured their position, or someone has taken their place, so there is no need for the higher levels of economic activity and subsequent pollution regulation violations.
